```python
import numpy as np

import sys
sys.path.insert(0, '../')
from chp3_fluxfuncs import get_nliw_times
from chp3_samplefuncs import sample_density, sample_celerity
```

```python
moor_file = 'T150'
field_trip = 'RS2019'
```

```python
trace_name = 'RS2019_T150_3days'
```

```python
nsamp = 10
ngrid = 50
rhoall, zout, time_np = sample_density(trace_name, gridpoints=None, samples=nsamp, noise=True)
```

```python
mode = 0
cel = sample_celerity(time_np, zout, rhoall, nsamp, mode=0)
```

```python
import matplotlib.pyplot as plt
plt.plot(time_np, cel.mean(axis=1))
plt.fill_between(time_np, np.percentile(cel, 0.05, axis=1), np.percentile(cel, 99.95, axis=1), alpha=0.5)
plt.ylim(0.0, 1.0)
plt.grid()
```

```python
# Load solibore wave times and print
wav_tx = get_nliw_times()
for xx, wx in enumerate(wav_tx):
    ix = np.argmin(np.abs(time_np - wx))
    print(f'Event {xx}: {cel[ix,:].mean():0.3f} ({np.percentile(cel[ix,:], 2.5):0.3f} {np.percentile(cel[ix,:], 97.5):0.3f})')
```
